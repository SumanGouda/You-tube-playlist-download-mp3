```python
url = "https://www.youtube.com/watch?v=7z2XJlKFmTo&list=PLFkNf3En4-HJBiFY-xY3QRW3W-n4pORig"
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get(url)
time.sleep(10) 

video_renderers = driver.find_elements(By.TAG_NAME, "ytd-playlist-panel-video-renderer")

# Open the file to dump the values
with open("song_link_1.txt", "w") as f:
    for renderer in video_renderers:
        try:
            # Use JavaScript to pierce the Shadow DOM and get the href
            link = driver.execute_script(
                "return arguments[0].querySelector('a#wc-endpoint').href;", 
                renderer
            )
            if link:
                # Write the link to the file with a newline
                f.write(link + "\n")
        except Exception as e:
            continue

driver.quit()
print("Links have been successfully dumped into song_link_1.txt")
```

```python
import yt_dlp
import os

# Create a folder for your MP3s
if not os.path.exists('downloads'):
    os.makedirs('downloads')

# Configuration for yt-dlp
ydl_opts = {
    'format': 'bestaudio/best',
    'noplaylist': True,
    
    # It tells YouTube we are a mobile device, which bypasses most bot checks.
    'user_agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 16_5 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/16.5 Mobile/15E148 Safari/604.1',
    
    # Your standard MP3 conversion settings
    'postprocessors': [{
        'key': 'FFmpegExtractAudio',
        'preferredcodec': 'mp3',
        'preferredquality': '192',
    }],
    'outtmpl': 'downloads/%(title)s.%(ext)s',
}

def download_as_mp3():
    # Read links from your saved file
    with open("song_link_1.txt", "r") as f:
        links = [line.strip() for line in f if line.strip()]

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        print(f"Starting download of {len(links)} songs...")
        ydl.download(links)

if __name__ == "__main__":
    download_as_mp3()
```
